# Popularity Base Recommendation System

```python
rating_with_name = rating.merge(book, on='ISBN')
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_Rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_Rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_Rating'] >= 250].sort_values('avg_Rating', ascending=False).head(50)
popular_df
```

```python
popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df=popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_Rating','avg_Rating']]
popular_df
```

# Collaborate Filtring Based Recommender System

```python
rating_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # Find the index of the book
    index = np.where(pt.index == book_name)[0][0]

    # Get similar items sorted by similarity score
    similar_items = sorted(
        list(enumerate(similarity_score[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:5]  # Skip the first one (it's the book itself)

    data = []
    for i in similar_items:
        # Filter books matching the title
        temp_df = book[book['Book-Title'] == pt.index[i[0]]].drop_duplicates('Book-Title')

        # Extract details
        item = [
            temp_df['Book-Title'].values[0],
            temp_df['Book-Author'].values[0],
            temp_df['Image-URL-M'].values[0]
        ]

        data.append(item)

    return data
```

```python
recommend('1984')
```
