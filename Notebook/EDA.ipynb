```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
# Clean the Age column: Keep ages between 5 and 100
users_cleaned = users.copy()
users_cleaned = users_cleaned[(users_cleaned['Age'] >= 5) & (users_cleaned['Age'] <= 100)]
```

```python
# Plot distribution of Age
plt.figure(figsize=(10, 5))
sns.histplot(users_cleaned['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of User Ages (Cleaned)')
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.tight_layout()
plt.show()
```

```python
# Plot distribution of Book Ratings
plt.figure(figsize=(10, 5))
sns.countplot(x='Book-Rating', data=ratings, palette='viridis')
plt.title('Distribution of Book Ratings')
plt.xlabel('Book Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
